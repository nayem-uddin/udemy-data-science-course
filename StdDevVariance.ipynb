# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()  # standard deviation
```

```python
incomes.var() # variance
```
